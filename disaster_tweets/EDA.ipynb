```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/train.csv')
```

```python
df.head(10)
```

```python
df.target.plot(kind='hist')
```

```python
words = []
def get_all_word(cell_string):
    words.extend(cell_string.split(' '))
    
df.text.apply(get_all_word);
print(f"unique words: {len(set(words))}")
```

```python
from collections import Counter
count_map = Counter(words)
```

```python
word_df = pd.DataFrame.from_dict(data=count_map, orient='index', columns=['count'])
word_df['word'] = word_df.index
word_df = word_df.set_index(np.arange(len(word_df)))
word_df = word_df[['word', 'count']]
```

```python
word_df.sort_values(by='count', ascending=False).head(50
                                                     )
```

```python
word_df.plot(kind='bar')
```
